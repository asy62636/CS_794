Altering tensors in place

```python
import math
a = torch.tensor([0, math.pi/4, math.pi/2, 3 * math.pi/4, math.pi])
print('a:')
print(a)
print(torch.sin(a))
print(a)

b = torch.tensor([0, math.pi/4, math.pi/2, 3 * math.pi/4, math.pi])
print('\nb:')
print(b)
print(torch.sin_(b))
print(b)
```

```python
a = torch.ones(2,2)
b = torch.rand(2,2)

print('a:')
print(a)
print(a.add(b))
print(a)
print(a.add_(b))
print(a)
print(b)
print(b.mul(b))
print(b)
print(b.mul_(b))
print(b)
```

```python
a = torch.rand(2,2)
b = torch.rand(2,2)
c = torch.zeros(2,2)
old_id = id(c)

print(c)
d = torch.matmul(a,b, out = c)
print(c)

assert c is d
assert id(c) == old_id

torch.rand(2,2, out = c)
print(c)
assert id(c) == old_id
print(c)
```
